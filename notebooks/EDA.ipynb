The grade data is saved to 'Dataset COH 20xx' files. We need to pull these in:

```python
# Create dfs for each sheet in the didactic grades file

# Load Didactic Grades (All Sheets)
#didactic_path = r"C:\Users\chris\Documents\PA_Data_Analysis\PA-Student-Performance-Analysis\data\Didactic_Grades_Cohort_20-23.xlsx"
#didactic_sheets = pd.read_excel(didactic_path, sheet_name=None)  # Load all sheets

# Check available sheets in Didactic file
print("Didactic Sheets:", list(didactic_sheets.keys()))

# Check available sheets in Didactic file
print("Didactic Sheets:", list(didactic_sheets.keys()))  # âœ… Prints sheet names to verify them

# Extract actual DataFrames instead of just sheet names
didactic_2020_df = didactic_sheets[list(didactic_sheets.keys())[0]]  # Extract DataFrame from dictionary
didactic_2021_df = didactic_sheets[list(didactic_sheets.keys())[2]]
didactic_2022_df = didactic_sheets[list(didactic_sheets.keys())[4]]
didactic_2023_df = didactic_sheets[list(didactic_sheets.keys())[6]]

# Create a dictionary with clear names
didactic_grades_dfs = {
    "Cohort_2020": didactic_2020_df,
    "Cohort_2021": didactic_2021_df,
    "Cohort_2022": didactic_2022_df,
    "Cohort_2023": didactic_2023_df,
}

# Add a "Cohort Year" column to each DataFrame
for cohort_name, df in didactic_grades_dfs.items():
    df["Cohort Year"] = cohort_name 
    print(f"Preview of {cohort_name}:")
    print(df.head(), "\n")

# Combine all cohort DataFrames into a single DataFrame
didactic_df = pd.concat(didactic_grades_dfs.values(), ignore_index=True)

# Preview the merged DataFrame
print(didactic_df.head())
print(didactic_df['Cohort Year'].unique())


```

Now that the datasets are aggregated, I want to run some analytics. 

Goals:
- Average score per class (per cohort/all cohorts)
- Class with lowest average score (per cohort/all cohorts)
- Class with highest average score (per cohort/all cohorts)
- Average GPA per semester (per cohort/all cohorts)
- Lowest GPA semester (all cohorts)
- 

```python
# Average score per class (per cohort/all cohorts)

# Average score per class (all cohorts)
avg_score_per_class = didactic_df.groupby("Course Name")["Grade"].mean().reset_index()
avg_score_per_class.rename(columns={"Grade": "Average Score"}, inplace=True)

# Average score per class (per cohort)
avg_score_per_class_cohort = didactic_df.groupby(["Cohort Year", "Course Name"])["Grade"].mean().reset_index()
avg_score_per_class_cohort.rename(columns={"Grade": "Average Score"}, inplace=True)

print("Average Score Per Class (All Cohorts):")
print(avg_score_per_class.head(21))

print("\n Average Score Per Class (Cohort 2020):")
print(avg_score_per_class_cohort.loc[avg_score_per_class_cohort["Cohort Year"] == "Cohort_2020"].head(21))

print("\n Average Score Per Class (Cohort 2021):")
print(avg_score_per_class_cohort.loc[avg_score_per_class_cohort["Cohort Year"] == "Cohort_2021"].head(21))

print("\n Average Score Per Class (Cohort 2022):")
print(avg_score_per_class_cohort.loc[avg_score_per_class_cohort["Cohort Year"] == "Cohort_2022"].head(21))

print("\n Average Score Per Class (Cohort 2023):")
print(avg_score_per_class_cohort.loc[avg_score_per_class_cohort["Cohort Year"] == "Cohort_2023"].head(21))


```
